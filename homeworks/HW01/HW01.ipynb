1. Импорт библиотеки NumPy

```python
import numpy as np
```

2. Пример создания массивов

```python
#Создание одномернного массива с помощью np.array и списка
 
list_1 = [1, 2, 3, 4, 5]
arr_1 = np.array(list_1)

#Создание двумерного массива с помощью np.array и списка

list_2 = [[1, 2, 3], [4, 5, 6]]
arr_2 = np.array(list_2)

print(f'Пример одномерного массива:\n {arr_1}')

print(f'Пример двумерного массива:\n {arr_2}')
```

```python
#np.zeros - заполняет массив нулями. Задается длинна массива
arr_zeros1 = np.zeros(10)

#Можно создать двумернный массив нулей.
#Первое значение - количество строк. Второе - количество столбцов
arr_zeros2 = np.zeros((5, 2))

print(f'Пример одномерного массива нулей:\n {arr_zeros1}')
print(f'Пример двумерного массива нулей:\n {arr_zeros2}')
```

```python
#np.ones - заполняет массив единицами. Задается длинна массива
arr_ones1 = np.ones(10)

#Можно создать двумернный массив единиц.
#Первое значение - количество строк. Второе - количество столбцов
arr_ones2 = np.ones((2, 5))

print(f'Пример одномерного массива единиц:\n {arr_ones1}')
print(f'Пример двумерного массива единиц:\n {arr_ones2}')
```

```python
#Создает массив, заполненный значениями от 0 до 9
arr_range1 = np.arange(10)

#Можно также задать диапазон чисел (от 1 до 30), а также шаг между значеняими (5)
arr_range2 = np.arange(0, 30, 5)

#Функция reshape меняет форму массива, без изменения данных
arr_range3 = arr_range2.reshape(2, 3)

print(f'Пример одномернного массива, заполненный значениями от 0 до 9:\n {arr_range1}')
print(f'Пример одномерного массива, от 0 до 30 с шагом 5:\n {arr_range2}')
print(f'Пример двумерного массива, от 0 до 30, с шагом 5:\n {arr_range3}')
```

```python
#Функция np.eye() создает единичную матрицу. Значение в матрице - размерность матрицы (3x3) 
matrix1 = np.eye(3)

#Можно задать точную размерность матрицы. Первое значение - количество строк, второе - значение столбцов
matrix2 = np.eye(5, 6)

#Можно указать диагональный сдвиг
matrix3 = np.eye(5, k = 2)

print(f'Пример единичной матрицы:\n{matrix1}')
print(f'Пример единичной матрицы размерностью 5x6:\n{matrix2}')
print(f'Пример единичной матрицы с диагональным сдвигом:\n{matrix3}')
```

3. Демонстрация атрибутов

```python

arr_s1 = np.arange(1, 51)
arr_s2 = arr_s1.reshape(5, 10)
arr_s3 = np.ones((10, 10))
print(f'Размерность массива arr_s2: {arr_s2.shape}')
print(f'Размерность массива arr_s3: {arr_s3.shape}')

arr_type1 = np.full(10, 'Hello world')
arr_type2 = np.full(10, 5.0)
arr_type3 = np.full(10, 5)
print(f'Тип данных массива arr_type1: {arr_type1.dtype}')
print(f'Тип данных массива arr_type2: {arr_type2.dtype}')
print(f'Тип данных массива arr_type3: {arr_type3.dtype}')

#Тип данных массива '<U11' означает: '<' - строка, 'U' - кодировка Unicode, '11' - максимальное количество символов в строке
```

4. Примеры индексирования и срезов

```python
#Получение отдельного элемента одномерного массива
print(f'Пятый элемент массива arr_range2: {arr_range2[4]}')

#Получение последнего элемента одномерного массива
print(f'Последний элемент массива arr_range2: {arr_range2[-1]}')

#Получение отдельного элемента двумерного массива
print(f'Элемент 2x3 двумернного массива arr_range2: {arr_range3[0, 2]}')
```

```python
arr_s4 = np.arange(20)
print(f'Элементы массива начиная с 5-го индекса: {arr_s4[5:]}')
print(f'Элементы массива до 9-го индекса: {arr_s4[:10]}')
print(f'Элементы массива с шагом 5: {arr_s4[::5]}')
print(f'Элементы массива в обратном порядке: {arr_s4[::-1]}')
print(f'Элементы массива начиная с 0-го индекса и заканчивая 10-ым: {arr_s4[1:11]}\n')

print(f'Срез двумерного массива:\n {arr_s2[:3, 3:]}') #Первое значение - первые 3 строки, второе значение - последние 7 столбцов
```

5. Примеры векторных операций и базовых статистик

```python
print(f'Одномерные массивы')
arr_a = np.arange(5, 11)
arr_b = np.arange(15, 21)
print(f'Пожлементное сложение одномерных массивов (a+b):\n{arr_a + arr_b}')
print(f'Поэлементное умножение одномерных массивов (a*b):\n{arr_a * arr_b}')
print(f'Поэлементное умножение одномерных массивов (b/a):\n{arr_b / arr_a}')

print(f'\nДвумерные массивы')
arr_a2 = np.arange(9, 21).reshape(4, 3)
arr_b2 = np.arange(21, 33).reshape(4, 3)
print(f'Поэлементное сложение двумерных массивов (a+b):\n{arr_a2 + arr_b2}')
print(f'Поэлементное умножение двумерных массивов (a*b):\n{arr_a2 * arr_b2}')
print(f'Поэлементное умножение одномерных массивов (b/a):\n{arr_b2 / arr_a2}')
```

```python
data = np.arange(1, 51, 5)

print(data)

print(f'Среднее: {np.mean(data)}')
print(f'Медиана: {np.median(data)}')
print(f'Стандартное отклонение: {np.std(data)}')
print(f'Минимальное и максимальное значение:\n{np.min(data)} \n{np.max(data)}')
```
